# What is Digital Audio?

First of all, audio is a type of signal, similar to visible and invisible light. As humans, we can perceive audio signals with frequencies ranging from 20 Hz to 20,000 Hz.

Digital audio, on the other hand, represents audio information in a format that computers can comprehend. It is essentially a collection of numerical data points, forming a graph or waveform.
